# Things to know about plots

- two things we plot: data, theory (model)
- data: measured
- theory (model): trendline fit, curve fit, simulation, equation
- data: **always** use markers
- time-series data: can also use a line in addition to the markers
- theory: **always** use lines, **never** use markers
- make the plot as large as you can
- never ever ever have a legend outside of the plot

- fonts: on plots, **always** use a sans-serif font: Ariel, Helvetica, Calibri
- fonts: on running text (e.g. paper) **always** use a serif font: Times, Palatino, Cambria
