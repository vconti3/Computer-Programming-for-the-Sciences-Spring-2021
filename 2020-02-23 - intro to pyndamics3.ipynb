```python
from pyndamics3 import Simulation
```

$$
\frac{dp}{dt} = a\cdot p
$$

$$
\dot{p} = a\cdot p
$$


$$
p' = a\cdot p
$$

```python
sim=Simulation()
sim.figsize=(5,3)
sim.add(" p'=a*p ",2,plot=True)  # initial condition for p. what is p at t=0?
sim.params(a=0.5)
sim.run(10)  # run for the specified time
```

```python
sim=Simulation()
sim.figsize=(5,3)
sim.add(" p'=a*p ",2,plot=True)  # initial condition for p. what is p at t=0?
sim.add(" p2'=a2*p2 ",2)  # initial condition for p. what is p at t=0?
sim.params(a=0.5,a2=1)
sim.run(10)  # run for the specified time,
```

```python
sim=Simulation()
sim.figsize=(5,3)
sim.add(" p'=a*p ",2,plot=1)  # initial condition for p. what is p at t=0?
sim.add(" p2'=a2*p2 ",2,plot=1)  # initial condition for p. what is p at t=0?
sim.params(a=0.5,a2=.6)
sim.run(10)  # run for the specified time,
```

$$
\frac{dv}{dt}=g-k\cdot v^2/m

Fixed point for $v$?


$$
\frac{dv}{dt}=0
$$

$$
v = \sqrt{mg/k}
$$
